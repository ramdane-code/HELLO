```python
import numpy as np 
import pandas as pd 

vtes_cream = np.array ([30,40,35,130,120,60])
code_meteo = np.array([0,1,0,1,1,0])
nbre_clients = np.array([2000,2100,1500,8000,7200,2000])

df = pd.DataFrame({"vtes": vtes_cream, "code": code_meteo, "clients" : nbre_clients})
print(f"Tab1 : \n{df}")
```

```python
monIndex = pd.date_range(start="04/2010", freq="ME", periods=6)
df.set_index(monIndex,inplace=True)
print(f"Tab2 : \n{df}")
```

```python
df["vtes_lag"]=df["vtes"].shift(2)
print(f"Tab3 : \n{df}")
```

```python
df["meteo"]= pd.Categorical.from_codes(codes=df["code"], categories=["mauvais","bon"])
print(f"Tab4: \n{df}")
```

```python
group_mean= df.groupby("meteo").mean()
print(f"Tab5 : \n{group_mean}")
```
